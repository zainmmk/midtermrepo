```python
weather_key = os.environ["WETH_API"]
```

```python
jsonlist = []
for ele in city_names_sample:
    i=0
    url = 'http://api.weatherapi.com/v1/current.json'
    params = {'key': weather_key, 'q':ele, 'dt':date_flight_sample[i]}
    response = re.get(url, params = params)
    i+=1
    b = response.json()
    print(response.status_code)
    jsonlist.append(b)
```

```python
for ele in jsonlist:
    print(ele['current']['condition']['text'])
```
